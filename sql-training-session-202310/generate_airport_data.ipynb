```python
from faker import Faker
from faker import providers
from faker.providers import DynamicProvider
from random import randint
import pandas as pd
```

```python
# define list of airports for provider
airports = [
    "SFO", "MSP", "DET", "LAX", "OAK", "LAS", "ORD", "JFK", "LGA", "DCA", "ORD", "ATL"
]

flight_times = ["AM", "AM - Early", "PM", "PM - Late", "Red Eye"]
```

```python
# define faker (with seed) to create dataset of
# date of travel
# passenger name
# passenger age
# departure airport
# destination airport
# travel window
faker = Faker('en_US')
Faker.seed(100)

airport_provider = DynamicProvider(
     provider_name="airport", elements=airports,
)
flight_time_provider = DynamicProvider(
    provider_name="flight_times", elements=flight_times
)

# add providers to faker
faker.add_provider(airport_provider)
faker.add_provider(flight_time_provider)
```

```python
# generate fake records
fakers = []
for i in range(150000):
    record = dict()
    record["travel_date"] = faker.date_this_month()
    record["passenger_name"] = faker.name()
    record["passenger_age"] = randint(5, 68)
    record["departure_airport"] = faker.airport()
    record["destination_airport"] = faker.airport()
    record["travel_window"] = faker.flight_times()
    fakers.append(record)

# convert into dataframe and preview
airport_data = pd.DataFrame(fakers)
airport_data.head()
```

```python
# look at size, but we want to filter out data where departure airport = destination airport
airport_data.info()
```

```python
# filter out data where departure airport = destination airport
airport_data_filtered = airport_data[airport_data["departure_airport"] != airport_data["destination_airport"]]
airport_data_filtered.info()
```

```python
# we have enough!! saving
airport_data_filtered.to_csv("airport_data.csv", index=False)
```
